## Group 1

```python
def f1 (n):
    return n ** 0.999999 * np.log(n)

def f2 (n):
    return 1000000 * n

def f3 (n):
    return 1.000001 ** n

def f4 (n):
    return n ** 2
```

```python
n_values = np.linspace(2 , 1000, 500)

f1_values = f1(n_values)
f2_values = f2(n_values)
f3_values = f3(n_values)
f4_values = f4(n_values)
```

```python
plt.figure(figsize=(10,6))
plt.plot(n_values, f1_values, color = 'red', label = 'f1')
#plt.plot(n_values, f2_values, color = 'green', label = 'f2')
plt.plot(n_values, f3_values, color = 'yellow', label = 'f3')
#plt.plot(n_values, f4_values, color = 'purple', label = 'f4')

plt.legend()
plt.grid(True)
plt.title('Estimating Big-O Complexity within various functions')
plt.show()
```

* f1 function
* f4 function
* f2 function

## Merge Sort VS Insertion Sort Complexity Functions (LEC 3)

```python
def f1 (n):
    return 2.2 * n * np.log(n)

def f2 (n):
    return 0.2 * np.power(n,2)
```

```python
n_values = np.linspace(2 , 10000, 500)

f1_values = f1(n_values)
f2_values = f2(n_values)

```

```python
plt.figure(figsize=(10,6))
plt.plot(n_values, f1_values, color = 'red', label = 'Merge Sort')
plt.plot(n_values, f2_values, color = 'green', label = 'Insertion Sort')


plt.legend()
plt.grid(True)
plt.title('Estimating Big-O Complexity of Merge and Insertion Sorting Functions')
plt.xlabel('Input(n)')
plt.ylabel('Response')
plt.show()

```
